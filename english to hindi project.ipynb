# Data Preperation

```python
# Function to clean the text
def clean_text(text):
    text = text.lower()
    text = re.sub(r"i'm", "i am", text)
    text = re.sub(r"he's", "he is", text)
    text = re.sub(r"she's", "she is", text)
    text = re.sub(r"that's", "that is", text)
    text = re.sub(r"what's", "what is", text)
    text = re.sub(r"where's", "where is", text)
    text = re.sub(r"let's", "let us", text)
    text = re.sub(r"one's", "one is", text)
    text = re.sub(r"nobody's", "nobody is", text)
    text = re.sub(r"who's", "who is", text)
    text = re.sub(r"it's", "it is", text)
    text = re.sub(r"how's", "how is", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"isn't", "is not", text)
    text = re.sub(r"won't", "will not", text)
    text = re.sub(r"can't", "cannot", text)
    text = re.sub(r"don't", "do not", text)
    text = re.sub(r"didn't", "did not", text)
    text = re.sub(r"couldn't", "could not", text)
    text = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,]", "", text)
    text=re.sub('\n','',text)
    return text
```

```python
# Cleaning the english sentences 
clean_eng=[]
for line in english:
    clean_eng.append(clean_text(line))
    
    
```

```python
# Function to clean hindi text
def clean_hind(text):
    text=re.sub(r'\n','',text)
    text=re.sub(r'-',' ',text)
    text=re.sub(r'[!.?।]','',text)
    return text
```

```python
# Cleaning the hindi text
clean_hindi=[]
for line in hindi:
    clean_hindi.append(clean_hind(line))
    
```

# Converting data into the required format for seq2seq(by tokenizng and padding)

```python
# encoder_input_data
tokenized_questions = tokenizer_eng.texts_to_sequences( clean_eng )
maxlen_questions = max( [ len(x) for x in tokenized_questions ] )
padded_questions = preprocessing.sequence.pad_sequences( tokenized_questions , maxlen=maxlen_questions ,padding='post')
encoder_input_data = np.array( padded_questions )
print( encoder_input_data.shape , maxlen_questions )
```

```python
# decoder_input_data
tokenized_answers = tokenizer_hindi.texts_to_sequences( hindi_sent )
maxlen_answers = max( [ len(x) for x in tokenized_answers ] )
padded_answers = preprocessing.sequence.pad_sequences( tokenized_answers , maxlen=maxlen_answers,padding='post'  )
decoder_input_data = np.array( padded_answers )
print( decoder_input_data.shape , maxlen_answers )
```

```python
# decoder_output_data
tokenized_answers = tokenizer_hindi.texts_to_sequences( hindi_sent )
for i in range(len(tokenized_answers)) :
    tokenized_answers[i] = tokenized_answers[i][1:]
padded_answers = preprocessing.sequence.pad_sequences( tokenized_answers , maxlen=maxlen_answers,padding='post' )
decoder_output_data = np.array(padded_answers)
decoder_output_data=decoder_output_data.reshape((decoder_output_data.shape[0],decoder_output_data.shape[1],1))
print( decoder_output_data.shape )
```

# Getting word embeddings

```python
# Getting english embeddings from glove2vec pretrained with 300 dimensions
embeddings_dict = {}
with open("Downloads/glove.6B.300d.txt", 'r', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        token = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[token] = vector
```

```python
embedding_dim=300
```

```python
num_words=len(word_index_english)+1
embedding_matrix=np.zeros((num_words,embedding_dim))
for word,i in word_index_english.items():
    if i>num_words:
        continue
    embedding_vector=embeddings_dict.get(word)
    if embedding_vector is not None:
        embedding_matrix[i]=embedding_vector
```

```python
# Getting hindi embeddings from fasttext pretrained with 300 dimensions
embeddings_dict_hindi = {}
i=0
with open("C:/Users/ASUS/Downloads/cc.hi.300.vec", 'r', encoding='utf-8', newline='\n', errors='ignore') as f:
    for line in f:
        values = line.split()
        token = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict_hindi[token] = vector
        i+=1
        if i==400000:
            break
```

```python
num_words_hindi=len(word_index_hindi)+1
embedding_matrix_hindi=np.zeros((num_words_hindi,embedding_dim))
for word,i in word_index_hindi.items():
    if i>num_words:
        continue
    embedding_vector=embeddings_dict_hindi.get(word)
    if embedding_vector is not None:
        embedding_matrix_hindi[i]=embedding_vector
```

# Defining the model for application 

```python
#Defining model architecture for seq2seq Model

encoder_inputs = tf.keras.layers.Input(shape=( None , ))
encoder_embedding = tf.keras.layers.Embedding( num_words, 300 , mask_zero=True,weights=[embedding_matrix],trainable=False ) (encoder_inputs)
encoder_outputs , state_h , state_c = tf.keras.layers.LSTM( 300 , return_state=True )( encoder_embedding )
encoder_states = [ state_h , state_c ]

decoder_inputs = tf.keras.layers.Input(shape=( None ,  ))
decoder_embedding = tf.keras.layers.Embedding( num_words_hindi, 300 , mask_zero=True,weights=[embedding_matrix_hindi],trainable=False) (decoder_inputs)
decoder_lstm = tf.keras.layers.LSTM( 300 , return_state=True , return_sequences=True )
decoder_outputs , _ , _ = decoder_lstm ( decoder_embedding , initial_state=encoder_states )
decoder_dense = tf.keras.layers.Dense( num_words_hindi , activation=tf.keras.activations.softmax ) 
output = decoder_dense( decoder_outputs)

model = tf.keras.models.Model([encoder_inputs, decoder_inputs], output )
model.compile(optimizer=tf.keras.optimizers.Adam(), loss='sparse_categorical_crossentropy',metrics=['sparse_categorical_accuracy'])

model.summary()
```

```python
# Running the model till 50 epochs
model.fit([encoder_input_data , decoder_input_data], decoder_output_data, batch_size=16, epochs=50,verbose=2) 
```

# Creating an inference model 

```python
# Defining the model with encoder and decoder states

def make_inference_models():
    
    encoder_model = tf.keras.models.Model(encoder_inputs, encoder_states)
    
    decoder_state_input_h = tf.keras.layers.Input(shape=( 300 ,))
    decoder_state_input_c = tf.keras.layers.Input(shape=( 300 ,))
    
    decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]
    
    decoder_outputs, state_h, state_c = decoder_lstm(
        decoder_embedding , initial_state=decoder_states_inputs)
    decoder_states = [state_h, state_c]
    decoder_outputs = decoder_dense(decoder_outputs)
    decoder_model = tf.keras.models.Model(
        [decoder_inputs] + decoder_states_inputs,
        [decoder_outputs] + decoder_states)
    
    return encoder_model , decoder_model
```

```python
# tokenizing the input sentences and then padding them

def str_to_tokens( sentence : str ):
    words = sentence.lower().split()
    tokens_list = list()
    for word in words:
        tokens_list.append( tokenizer_eng.word_index[ word ] ) 
    return preprocessing.sequence.pad_sequences( [tokens_list] , maxlen=maxlen_questions,padding='post' )
```
